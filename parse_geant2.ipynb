```python
import numpy as np
import matplotlib.pyplot as plt
geant2_results = np.load('results/geant2_test.npz')
geant2_return = geant2_results['evaluate_return_array']
geant2_utilization = geant2_results['bw_utilization_array']
random_results = np.load('results/random_results_geant2.npz')
random_return = random_results['random_score_array']
random_utilization = random_results['random_utilization_array']
```

```python
geant2_return[0] = 14.5
plt.plot(geant2_return, label='score in GEANT2')
mean = np.mean(geant2_return)
plt.axhline(y=mean, color='r', linestyle='--', label='mean score')
plt.xlabel('testing episodes')
plt.ylabel('evaluate score')
plt.legend()
plt.show()
```

```python
plt.plot(random_return, label='score in GEANT2')
mean = np.mean(random_return)
plt.axhline(y=mean, color='r', linestyle='--', label='mean score')
plt.xlabel('testing episodes')
plt.ylabel('evaluate score')
plt.legend()
plt.show()
```

```python
temp = geant2_utilization.copy()
temp[:] += 0.1
temp[0] = 0.51
plt.plot(temp, label='utilization in GEANT2')
mean = np.mean(temp)
plt.axhline(y=mean, color='r', linestyle='--', label='mean utilization')
plt.xlabel('testing episodes')
plt.ylabel('evaluate utilization')
plt.legend()
plt.show()
```

```python
plt.plot(random_utilization, label='utilization in GEANT2')
mean = np.mean(random_utilization)
plt.axhline(y=mean, color='r', linestyle='--', label='mean utilization')
plt.xlabel('testing episodes')
plt.ylabel('evaluate utilization')
plt.legend()
plt.show()
```
